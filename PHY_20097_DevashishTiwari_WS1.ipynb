# Now, As We have Extracted the Web Links for each page, so, Let us now Extract the Information from each Page seperately

# For my understanding, I am defining the Elements again, and scraping each link, then we will run the loop

```python
urls = []

for i in range(1, 12):
    if i == 11 or i == 12 or i == 10:
        ur = f"https://arxiv.org/list/math-ph/22{i}"
        print(ur)
        
    else:
        ur = f"https://arxiv.org/list/math-ph/220{i}"
        print(ur)
        
    urls.append(ur)
```

```python
#As, We are doing again and again, for getting the info regarding the Wrbpage, so let usNow, Define the functions, and use that
#Again and again, so as to avoids confusions
```

```python
def info_paper(url_paper):
    response=requests.get(url_paper,headers=headers)
    
    if response.status_code != 200:
        print('Status code:', response.status_code)
        raise Exception('Failed to fetch web page ')

    doc_paper = BeautifulSoup(response.text,'html.parser')

    return doc_paper
```

```python
#Obtaining the Name of Paper from each websites, thus I have used For loop within the loop
```

```python
df_name = []

for i in range(0, 8):
    docs = info_paper(urls[i])
    
    nice = docs.find_all('div',{"class" : "meta"})

    
    data = []
    for y in range(len(nice)):
        for div in nice[y].find_all('div', "list-title mathjax"):
            for tag in div.children:
                a = tag.text
                #print(a)
            
        d = {
            'Name of the Research Paper' : a
        }
    
        data.append(d)
        
    df1 = pd.DataFrame(data)
    #print(df1)
    
    folder_path = f"C:\\Users\\Devashish\\Dropbox\\PC\\Desktop\\World of Programming\\Python\\DSP\\Web Scrapping\\Data\\Name of Paper\\ResearchPaperName{i}.csv"
    df1.to_csv(folder_path, index=False)
    
    df_name.append(df1)
    
    
df_merged1 = pd.concat(df_name)
print(df_merged1)
```

```python
#Obtaining the Authors from each websites, thus I have used For loop within the loop
```

```python
df_authors = []
datam = []

for i in range(0, len(urls)):
    docs = info_paper(urls[i])
    
               
    nice = docs.find_all('div',{"class" : "meta"})

    k = []
    l = []

    for y in range(0, len(nice)):
               a = nice[y].find_all('div', "list-authors")[0]
               b = a.text
               #print(b)
               k.append(b)
     
    for j in range(0, 25):
               c = k[j].split()
               m = c[1:][1]
               #print(m)
    
               l.append(m)
        
    d = {
        'Authors' : l
        }
    
    datam.append(d)
    
    l
    df3 = pd.DataFrame(d)
    print(df3)
    
    folder_path = f"C:\\Users\\Devashish\\Dropbox\\PC\\Desktop\\World of Programming\\Python\\DSP\\Web Scrapping\\Data\\Authors\\Authors{i}.csv"
    df3.to_csv(folder_path, index=False)
```

```python
#Obtaining the subjects from each websites, thus I have used For loop within the loop
```

```python
df_list_subjects = []

for i in range(0, len(urls)):
    docs = info_paper(urls[i])
    
    nice = docs.find_all('div',{"class" : "meta"})

    data2 = []

    for i in range(len(nice)):
        for div in nice[i].find_all('div', "list-subjects"):
            for tag in div.children:
                a = tag.text
                
                if a == "\n":
                    p = "Mathematical Physics"
                else:
                    p = a
                    #print(a)
            
        d = {
            'Subjects' : p
             }
    
        data2.append(d)
    
    data2
    df2 = pd.DataFrame(data2)
    #print(df2)
    
    folder_path = f"C:\\Users\\Devashish\\Dropbox\\PC\\Desktop\\World of Programming\\Python\\DSP\\Web Scrapping\\Data\\Subjects\\SubjectResearchPaper{i}.csv"
    df2.to_csv(folder_path, index=False)
    df_list_subjects.append(df2)

df_merged3 = pd.concat(df_list_subjects)
print(df_merged3)
```
